```python
import matplotlib.pyplot as plt
import numpy as np


train_loss_4096 = {"gpt": 4.1576, "kda":7.0288, "Hybrid":7.0172}
test_loss_4096 = {"gpt": 5.0808, "kda":6.6250, "Hybrid":6.6680}

train_loss_1024 = {"gpt":3.3495, "kda":7.0531, "Hybrid":7.0409}
test_loss_1024 = {"gpt":4.8225, "kda":6.6361, "Hybrid":6.6233}

labels = ['gpt', 'kda', 'hybrid']
x = np.arange(len(labels))
train_1024_values = list(train_loss_1024.values())
train_4096_values = list(train_loss_4096.values())
test_1024_values = list(test_loss_1024.values())
test_4096_values = list(test_loss_4096.values())
print(train_1024_values)
fig, axes = plt.subplots(1, 2, figsize=(14, 7))

def annotate_points(ax, x, y, offset_y=12, color='#333333'):
    for xi, yi in zip(x, y):
        ax.annotate(f'{yi:.4f}', (xi, yi), textcoords="offset points", # 使用"偏移像素"坐标系
                   xytext=(0, offset_y), # 文字相对于点的偏移量
                   ha='center', fontsize=9, color=color)


ax1 = axes[0]
ax1.plot(x, train_1024_values, 'o-', label='train_loss_1024', linewidth=2.5, markersize=10, color='#1f77b4')
ax1.plot(x, train_4096_values, 's-', label='train_loss_4096', linewidth=2.5, markersize=10, color='#ff7f0e')

annotate_points(ax1, x, train_1024_values, offset_y=12, color='#1f77b4')
annotate_points(ax1, x, train_4096_values, offset_y=-18, color='#ff7f0e')
ax1.set_xticks(x)
ax1.set_xticklabels(labels)
ax1.set_xlim(-0.5, 2.5)
ax1.set_ylim(3, 7.5)  
ax1.set_xlabel('Model', fontsize=12)
ax1.set_ylabel('Loss', fontsize=12)
ax1.set_title('Train Loss Comparison', fontsize=14, fontweight='bold')
ax1.legend(loc='best', fontsize=10)
ax1.grid(True, alpha=0.3, linestyle='--')


ax2 = axes[1]
ax2.plot(x, test_1024_values, 'o-', label='test_loss_1024', linewidth=2.5, markersize=10, color='#1f77b4')
ax2.plot(x, test_4096_values, 's-', label='test_loss_4096', linewidth=2.5, markersize=10, color='#ff7f0e')

annotate_points(ax2, x, test_1024_values, offset_y=-18, color='#1f77b4')
annotate_points(ax2, x, test_4096_values, offset_y=18, color='#ff7f0e')
ax2.set_xticks(x)
ax2.set_xticklabels(labels)
ax2.set_xlim(-0.5, 2.5)
ax2.set_ylim(3, 7.5)  
ax2.set_xlabel('Model', fontsize=12)
ax2.set_ylabel('Loss', fontsize=12)
ax2.set_title('Test Loss Comparison', fontsize=14, fontweight='bold')
ax2.legend(loc='best', fontsize=10)
ax2.grid(True, alpha=0.3, linestyle='--')

plt.tight_layout()
plt.show()
```
